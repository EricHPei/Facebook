```python
top_ten_places = dfsample['place_id'].value_counts()[:9]
plt.figure(4, figsize=(14,15))
for index, value in enumerate(top_ten_places.index):
    dfsample_sub = dfsample[dfsample.loc[:,'place_id'] == value]
    x = dfsample_sub['time']
    y = dfsample_sub['x']
    graph_loc = int(str('33')+str(index+1))
    plt.subplot(graph_loc)
    gridsize=50
    plt.title(value)
    plt.xlabel('Time')
    plt.ylabel('X')
    plt.hexbin(x, y, gridsize=gridsize, cmap=CM.jet, bins=None)
    plt.axis([x.min(), x.max(), y.min(), y.max()])
    
plt.show()
```
